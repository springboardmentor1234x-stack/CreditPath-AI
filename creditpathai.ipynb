```python
# ======================================
# 🔹 STEP 1: Import libraries
# ======================================
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# ML Models
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# ======================================
# 🔹 STEP 2: Load dataset
# ======================================
uploaded = files.upload()  # Upload your Loan_Default.csv
df = pd.read_csv("Loan_Default.csv")
print("✅ Data Loaded Successfully")
print("Shape:", df.shape)
display(df.head())

# ===============================================================
# 📊 STEP 3: Exploratory Data Analysis (EDA)
# ===============================================================

# 🔹 Distribution of Loan Amount
if 'loan_amount' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['loan_amount'], bins=30, kde=True, color='teal')
    plt.title("Distribution of Loan Amount")
    plt.xlabel("Loan Amount")
    plt.ylabel("Frequency")
    plt.show()

# 🔹 Correlation Heatmap
# Calculate correlation only on numerical columns
numerical_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(10,6))
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

# 🔹 Loan Status Distribution
if 'Status' in df.columns:
    plt.figure(figsize=(6,4))
    df['Status'].value_counts().plot(kind='bar', color=['green','red'])
    plt.title("Loan Status Distribution (0 = Non-Default, 1 = Default)")
    plt.xlabel("Status")
    plt.ylabel("Count")
    plt.show()

# 🔹 Summary Statistics
print("📈 Summary Statistics for Results DataFrame:")
# Check if results_df exists before trying to display it
if 'results_df' in locals():
    display(results_df.describe(include='all'))
else:
    print("Results DataFrame not yet created.")


print("✅ EDA Completed Successfully!")


# ======================================
# 🔹 STEP 4: Handle missing values
# ======================================
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("✅ Missing values handled!")

# ======================================
# 🔹 STEP 5: Encode categorical columns
# ======================================
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
print("✅ Label Encoding Done!")

# ======================================
# 🔹 STEP 6: Remove duplicates and outliers
# ======================================
df.drop_duplicates(inplace=True)

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
print("✅ Outliers removed!")

# ======================================
# 🔹 STEP 7: Feature Scaling
# ======================================
# Exclude 'Status' column from numerical scaling
num_cols_to_scale = [col for col in num_cols if col != 'Status']
scaler = StandardScaler()
df[num_cols_to_scale] = scaler.fit_transform(df[num_cols_to_scale])
print("✅ Feature Scaling Done!")

# ======================================
# 🔹 STEP 8: Define X and y, and handle imbalance using SMOTE
# ======================================
target = "Status"  # Change if your target column name is different
if target in df.columns:
    X = df.drop(target, axis=1)
    y = df[target]

    sm = SMOTE(random_state=42)
    X_res, y_res = sm.fit_resample(X, y)
    print("✅ Class imbalance handled using SMOTE!")
else:
    print("⚠️ Target column not found.")

# ======================================
# 🔹 STEP 9: Train-Test Split
# ======================================
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
print("✅ Data Split into Train & Test Sets")

# ======================================
# 🔹 STEP 10: Train Models
# ======================================

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

# XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)

# LightGBM
lgb = LGBMClassifier(random_state=42)
lgb.fit(X_train, y_train)
y_pred_lgb = lgb.predict(X_test)

# ======================================
# 🔹 STEP 11: Evaluate Models
# ======================================
def evaluate_model(name, y_test, y_pred):
    print(f"\n📊 Model: {name}")
    print("Accuracy:", round(accuracy_score(y_test, y_pred)*100, 2), "%")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

evaluate_model("Logistic Regression", y_test, y_pred_lr)
evaluate_model("XGBoost", y_test, y_pred_xgb)
evaluate_model("LightGBM", y_test, y_pred_lgb)

print("🎯 All models trained and evaluated successfully!")

# ======================================
# 🔹 STEP 12: Hyperparameter Tuning
# ======================================
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

# Logistic Regression Hyperparameters
lr_params = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=1000), lr_params, cv=3, n_jobs=-1)
grid_lr.fit(X_train, y_train)
best_lr = grid_lr.best_estimator_
print("✅ Best Logistic Regression Params:", grid_lr.best_params_)

# XGBoost Hyperparameters
xgb_params = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1]
}

random_xgb = RandomizedSearchCV(
    XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    xgb_params, n_iter=10, cv=3, n_jobs=-1, random_state=42)
random_xgb.fit(X_train, y_train)
best_xgb = random_xgb.best_estimator_
print("✅ Best XGBoost Params:", random_xgb.best_params_)

# LightGBM Hyperparameters
lgb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'num_leaves': [31, 50, 70],
    'boosting_type': ['gbdt', 'dart']
}

random_lgb = RandomizedSearchCV(
    LGBMClassifier(random_state=42),
    lgb_params, n_iter=10, cv=3, n_jobs=-1, random_state=42)
random_lgb.fit(X_train, y_train)
best_lgb = random_lgb.best_estimator_
print("✅ Best LightGBM Params:", random_lgb.best_params_)

# ======================================
# 🔹 STEP 13: Evaluate Tuned Models
# ======================================
models = {
    'Logistic Regression': best_lr,
    'XGBoost': best_xgb,
    'LightGBM': best_lgb
}

results = {}

for name, model in models.items():
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"\n📊 Model: {name}")
    print("Accuracy:", round(acc*100, 2), "%")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Greens')
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

# ======================================
# 🔹 STEP 14: Compare Model Performance
# ======================================
plt.figure(figsize=(8,5))
sns.barplot(x=list(results.keys()), y=list(results.values()))
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.ylim(0, 1)
for i, v in enumerate(results.values()):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center', fontweight='bold')
plt.show()

print("🎯 Best Performing Model:", max(results, key=results.get))

# ======================================
# 🔹 STEP 15: Model Calibration
# ======================================
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import roc_auc_score, roc_curve

# Use the best model from previous step
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"✅ Using Best Model for Calibration: {best_model_name}")

# Calibrate model probabilities
calibrated_model = CalibratedClassifierCV(best_model, method='sigmoid', cv=3)
calibrated_model.fit(X_train, y_train)

# Predict probabilities
y_prob = calibrated_model.predict_proba(X_test)[:, 1]

# ======================================
# 🔹 STEP 16: Evaluate Calibration
# ======================================
prob_true, prob_pred = calibration_curve(y_test, y_prob, n_bins=10)

plt.figure(figsize=(6,6))
plt.plot(prob_pred, prob_true, marker='o', linewidth=1)
plt.plot([0,1],[0,1],'--', color='gray')
plt.title(f"Calibration Curve for {best_model_name}")
plt.xlabel("Predicted Probability")
plt.ylabel("True Probability")
plt.show()

roc_auc = roc_auc_score(y_test, y_prob)
print("📈 ROC-AUC Score:", round(roc_auc, 3))

# ======================================
# 🔹 STEP 17: Optimize Decision Threshold
# ======================================
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# You can pick threshold by maximizing (TPR - FPR)
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
print("✅ Optimal Threshold:", round(optimal_threshold, 3))

plt.figure(figsize=(7,5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.scatter(fpr[optimal_idx], tpr[optimal_idx], marker='o', color='red', label='Optimal Threshold')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.title(f"ROC Curve - {best_model_name}")
plt.show()

# ======================================
# 🔹 STEP 18: Apply Threshold & Create Action Segments
# ======================================
y_pred_custom = (y_prob >= optimal_threshold).astype(int)

# Add predictions to DataFrame
results_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted_Prob': y_prob,
    'Predicted_Label': y_pred_custom
})

# Segment customers
def assign_risk(prob):
    if prob >= 0.8:
        return 'High Risk - Reject / Investigate'
    elif prob >= 0.5:
        return 'Medium Risk - Manual Review'
    else:
        return 'Low Risk - Approve'

results_df['Action'] = results_df['Predicted_Prob'].apply(assign_risk)
display(results_df.head(10))

# ======================================
# 🔹 STEP 19: Visualize Risk Segments
# ======================================
risk_counts = results_df['Action'].value_counts()

plt.figure(figsize=(7,4))
sns.barplot(x=risk_counts.index, y=risk_counts.values)
plt.title("Loan Application Risk Segments")
plt.ylabel("Number of Applicants")
plt.xticks(rotation=15)
plt.show()

print("✅ Actionable insights generated successfully!")
# ======================================
# 🔹 STEP 20: Export Results to CSV and Excel
# ======================================
from google.colab import files

# Save to CSV
results_df.to_csv("Loan_Default_Actionable_Results.csv", index=False)
print("✅ CSV file created: Loan_Default_Actionable_Results.csv")

# Save to Excel (optional)
results_df.to_excel("Loan_Default_Actionable_Results.xlsx", index=False)
print("✅ Excel file created: Loan_Default_Actionable_Results.xlsx")

# Download files from Colab
files.download("Loan_Default_Actionable_Results.csv")
files.download("Loan_Default_Actionable_Results.xlsx")

print("📥 You can now download both CSV and Excel files.")
```
